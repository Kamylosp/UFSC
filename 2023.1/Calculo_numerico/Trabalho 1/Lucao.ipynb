```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im

c = 0.00261438
pos = [252, 44]

image = np.load('ufsc_j.npy')
imagefp_16 = np.load('ufsc_j.npy')
imagefp_64 = np.load('ufsc_j.npy')
diferenca_1000 = np.load('ufsc_j.npy')

plt.imshow(image)
plt.show()
```

```python
for a in range(0, len(imagefp_16)):  #
    for b in range(0, len(imagefp_16[a])):  # trabalha a imagem como se fosse uma matriz

        media_cores = sum(imagefp_16[a, b]) / 3  # soma o valor das cores e divide por 3
        media_cores = np.half(media_cores)  # transforma o valor em binary16

        imagefp_16[a, b] = 1*media_cores  # transforma todas as cores do pixel na media delas

im.imsave('image_fp16.jpg', imagefp_16)  # salva a imagem em binary16
print(imagefp_16[pos[0],pos[1]])

plt.imshow(imagefp_16)
plt.show()

plt.imshow(image)
plt.show()
```

```python
for a in range(0, len(imagefp_64)):               #
    for b in range(0, len(imagefp_64[a])):        #trabalha a imagem como se fosse uma matriz

        media_cores2 = sum(imagefp_64[a, b]) / 3   #soma o valor das cores e divide por 3

        imagefp_64[a, b] = 1*media_cores2       #transforma todas as cores do pixel na media delas

im.imsave('image_fp64.jpg', imagefp_64)      #salva a imagem em binary16
print(imagefp_64[pos[0],pos[1]])
plt.imshow(imagefp_64)
plt.show()

plt.imshow(image)
plt.show()
```

```python
inter = 0
inter2 = 0

for a in range(0, len(image)):  #
    for b in range(0, len(image[a])):  # trabalha a imagem como se fosse uma matriz

        inter = sum(imagefp_64[a, b]) / 3  # salva a media de cores da imagem em fp64
        inter2 = sum(imagefp_16[a, b]) / 3  # salva a media de cores da imagem em fp16

        inter = inter - inter2  # subtrai uma da outra
        inter = abs(inter)

        inter *= 1000  # multiplica por 1000 para ficar mais visível

        diferenca_1000[a, b] = inter, inter, inter  # adiciona a diferenca em um array

im.imsave('diferenca_1000.jpg', diferenca_1000)  # salva esse novo array como uma nova imagem
im.imsave('Imagemfinal.jpg', image)

plt.imshow(diferenca_1000)
plt.show()

plt.imshow(imagefp_16)
plt.show()

plt.imshow(imagefp_64)
plt.show()


plt.imshow(image)
plt.show()
```
