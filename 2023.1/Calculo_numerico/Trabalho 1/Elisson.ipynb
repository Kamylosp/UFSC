```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.image as im

imagem = mpimg.imread('Imagem1.png')[:]
print(imagem[100, 50])
print(imagem[100, 150])
print("cinza: ", imagem[118, 108])
print(imagem[10, 10])
for i in range(0, len(imagem)):
    for j in range(0, len(imagem[i])):
        if sum(imagem[i,j]) <= 0.01:
            imagem[i,j] = 1.0

im.imsave('Imagem1_modificada.png', imagem)
imgplot = plt.imshow(imagem)
plt.show()
```
