# Imagem

```python
import numpy as np
import matplotlib.pyplot as plt

const = 0.00392156862745
```

```python
image = np.load('ufsc_j.npy')
image2 = image[:]

cor_r, cor_g, cor_b = 0, 0, 0 # NÃºmero entre 0 e 255

for a in range(0, len(image2)):
    for b in range(0, len(image2[a])):
        # Alterando na imagem 2 e multiplicando pela const para o numero ficar entre 0 e 1
        image2[a,b] = const*cor_r, const*cor_g, const*cor_b
        
plt.imshow(image2)
```
